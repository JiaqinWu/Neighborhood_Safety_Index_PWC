```python
df = pd.read_csv('Race.csv')
df = df.transpose()
df.head()
```

```python
df1=df[[0,1,2,3,4,5,6,7,8]]
df1.head()
```

```python
# Set first row as header
df1.columns = df1.iloc[0]   
df1 = df1[1:]              
df2 = df1.reset_index(drop=True) 
df2.columns = df2.columns.str.strip()
df2 = df2.replace('[\$,%,]', '', regex=True)
df2 = df2.apply(pd.to_numeric, errors='coerce').astype('Int64')
df2.head()
```

```python
df2.columns
```

```python
df2['Prop_Hisp']  = df2['Hispanic or Latino']/df2['Total:']
df2['Prop_White'] = df2['White alone']/df2['Total:']
df2['Prop_Black'] = df2['Black or African American alone']/df2['Total:']
df2['Prop_American_Native'] = df2['American Indian and Alaska Native alone']/df2['Total:']
df2['Prop_Asian'] = df2['Asian alone'] /df2['Total:']
df2['Prop_Hawaiian'] = df2['Native Hawaiian and Other Pacific Islander alone']/df2['Total:']
df2['Prop_Onemorerace'] = (df2['Not Hispanic or Latino:']-df2['Population of one race:'])/df2['Total:']
df2.head()
```

```python
df2.to_csv('race_num.csv',index=False)
```
